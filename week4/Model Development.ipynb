# Simple Linear Regression

Simple linear regression or **SLR** is method to help us understand the reation between two variables. 
1. The predictor (independet) variable $X$
2. The target (dependent) variable $y$
### $$y={ b }_{ 0 }+{ b }_{ 1 }x$$
* ${ b }_{ 0 }$:The Intercept
* ${ b }_{ 1 }$:The Slope
